```python
X = df["question"]
y = df["intent"]


vectorizer = TfidfVectorizer(
    lowercase=True,
    stop_words="english"
)


X_vectorized = vectorizer.fit_transform(X)


print("Text processing completed")
```

```python
model = MultinomialNB()

model.fit(
    X_vectorized,
    y
)

print("AI model trained successfully")
```

```python
def chatbot_response(user_input):

    user_vector = vectorizer.transform([user_input])

    probabilities = model.predict_proba(user_vector)

    confidence = max(probabilities[0])

    predicted_intent = model.classes_[
        probabilities[0].argmax()
    ]

    if confidence < 0.20:
        return "Sorry, I could not understand your question. Please contact student support."

    response = df[
        df["intent"] == predicted_intent
    ]["answer"].iloc[0]

    return response
```

```python
import gradio as gr

def chat_function(message, history):

    response = chatbot_response(message)

    return response



demo = gr.ChatInterface(

    fn=chat_function,

    title="🎓 AI Student Support Chatbot",

    description="""
    Ask questions related to:
    - Admission
    - Fees
    - Exams
    - Scholarships
    - Library
    - Student Support
    """

)


demo.launch()
```
